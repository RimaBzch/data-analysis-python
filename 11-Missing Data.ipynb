# Missing Data in Pandas

```python
import pandas as pd
import numpy as np
```

```python
s=pd.Series(["Sam",np.nan,"Tim","Kim"])
s
```

```python
s.isnull()
```

```python
s.notnull()
```

```python
s[3]=None
s.isnull()
```

```python
s.dropna()
```

```python
from numpy import nan as NA
```

```python
df=pd.DataFrame([[1,2,3],[4,NA,5],
                 [NA,NA,NA]])
df
```

```python
df.dropna()
```

```python
df.dropna(how="all")
```

```python
df
```

```python
df.fillna(0)
```

```python
df
```

```python
df.fillna(0,inplace=True)
df
```

```python
df=pd.DataFrame([[1,2,3],[4,NA,5],
                 [NA,NA,NA]])
df
```

```python
df.fillna(method="ffill")
```
