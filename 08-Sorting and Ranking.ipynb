# Sorting & Ranking in Pandas

```python
import pandas as pd
import numpy as np
```

```python
s=pd.Series(range(5),
            index=["e","d","a","b","c"])
s
```

```python
s.sort_index()
```

```python
s.sort_values()
```

```python
df=pd.DataFrame(
    np.arange(12).reshape(3,4),
    index=["tawo","one","three"],
    columns=["d","a","b","c"])
df
```

```python
df.sort_index()
```

```python
df.sort_index(axis=1)
```

```python
df.sort_index(axis=1, ascending=False)
```

```python
s2=pd.Series([5,np.nan,3,-1,9])
s2
```

```python
s2.sort_values()
```

```python
df2=pd.DataFrame(
    {"a":["k","c0","f88","9","a"],"b":[1,1,-2,0,5]})
df2
```

```python
df2.sort_values(by="b")
```

```python
df2.sort_values(by=["b","a"])
```

```python
df2.sort_values(by="a")
```
