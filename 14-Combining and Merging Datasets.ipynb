# Combining & Merging Datasets in Pandas

```python
import pandas as pd
import numpy as np
```

## Joining DataFrame

```python
d1=pd.DataFrame(
    {"key":["a","b","c","c","d","e"],
     "num1":range(6)})
d2=pd.DataFrame(
    {"key":["b","c","e","f"],
     "num2":range(4)})
```

```python
print(d1)
print(d2)
```

```python
pd.merge(d1, d2)
```

```python
pd.merge(d1, d2, on='key')
```

```python
pd.merge(d1,d2,how="outer")
```

```python
pd.merge(d1,d2,how="left")
```

```python
pd.merge(d1,d2,how="right")
```

```python
pd.merge(d1, d2, how='inner')
```

## Concatenating Along an Axis

```python
seq= np.arange(20).reshape((4, 5))
```

```python
np.concatenate([seq,seq], axis=1)
```

```python
np.concatenate([seq, seq], axis=0)
```

```python
data1 = pd.Series(
    [0, 1], index=['a', 'b'])
data2 = pd.Series(
    [2,3,4], index=['c','d','e'])
data3 = pd.Series(
    [5, 6], index=['f', 'g'])
```

```python
pd.concat([data1,data2,data3])
```

```python
pd.concat([data1, data2, data3], axis=1)
```

```python
data1
```

```python
data4= pd.Series([10,11,12], 
                 index=['a','b',"c"])
pd.concat([data1,data4],axis=1,join="inner")
```

```python
df1 = pd.DataFrame(
    np.arange(6).reshape(3, 2),
    index=['a', 'b', 'c'],
    columns=['one', 'two'])
df2 = pd.DataFrame(
    10+np.arange(4).reshape(2,2),
    index=['a', 'c'],
    columns=['three', 'four'])
```

```python
pd.concat([df1, df2], axis=1, 
          keys=['s1', 's2'],
          sort=False)
```

```python
data1 = pd.DataFrame(
    np.random.randn(3, 4),
    columns=['a','b','c','d'])
data2 = pd.DataFrame(
    np.random.randn(2, 3),
    columns=['b','d','a'])
```

```python
pd.concat([data1, data2], ignore_index=True)
```
